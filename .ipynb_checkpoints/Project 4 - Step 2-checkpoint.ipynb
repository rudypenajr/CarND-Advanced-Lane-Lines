### HLS and Color Thresholds (Lesson 28)

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image = mpimg.imread('./test_images/color_space_test.jpg')
thresh = (180, 255)

gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
binary = np.zeros_like(gray)
binary[(gray > thresh[0]) & (gray <= thresh[1])] = 1


f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.imshow(gray, cmap="gray")
ax1.set_title('Gray', fontsize=30)
ax2.imshow(binary, cmap="gray")
ax2.set_title('Gray Binary', fontsize=30)
plt.show()
```

```python
thresh = (200, 255)
R = image[:,:,0]
G = image[:,:,1]
B = image[:,:,2]


f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))
ax1.imshow(R, cmap='gray')
ax1.set_title('R', fontsize=30)
ax2.imshow(G, cmap='gray')
ax2.set_title('G', fontsize=30)
ax3.imshow(B, cmap='gray')
ax3.set_title('B', fontsize=30)

plt.show()
```

The R channel does a reasonable job of highlighting the lines, and you can apply a similar threshold to find lane-line pixels, found below.

```python
binaryR = np.zeros_like(R)
binaryR[(R > thresh[0]) & (R <= thresh[1])] = 1

binaryG = np.zeros_like(G)
binaryG[(G > thresh[0]) & (G <= thresh[1])] = 1

binaryB = np.zeros_like(B)
binaryB[(B > thresh[0]) & (B <= thresh[1])] = 1

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))
ax1.imshow(binaryR, cmap='gray')
ax1.set_title('Binary R', fontsize=30)
ax2.imshow(binaryG, cmap='gray')
ax2.set_title('Binary G', fontsize=30)
ax3.imshow(binaryB, cmap='gray')
ax3.set_title('Binary B', fontsize=30)

plt.show()
```

In this lesson, we're looking at different color spaces. While there are several that are worth exploring, here we'll look specifically at HLS. When we separate the H, L, and S channels we get the following result:

```python
hls = cv2.cvtColor(image, cv2.COLOR_RGB2HLS)
H = hls[:,:,0]
L = hls[:,:,1]
S = hls[:,:,2]


f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))
ax1.imshow(H, cmap='gray')
ax1.set_title('H', fontsize=30)
ax2.imshow(L, cmap='gray')
ax2.set_title('L', fontsize=30)
ax3.imshow(S, cmap='gray')
ax3.set_title('S', fontsize=30)
 
plt.show()
```

The S channel picks up the lines well,s o let's try applying a threshold there:

```python
thresh = (90, 255)
binaryH = np.zeros_like(H)
binaryH[(H > thresh[0]) & (H <= thresh[1])] = 1
binaryL = np.zeros_like(H)
binaryL[(L > thresh[0]) & (L <= thresh[1])] = 1
binaryS = np.zeros_like(H)
binaryS[(S > thresh[0]) & (S <= thresh[1])] = 1


f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))
ax1.imshow(binaryH, cmap='gray')
ax1.set_title('Binary H', fontsize=30)
ax2.imshow(binaryL, cmap='gray')
ax2.set_title('Binary L', fontsize=30)
ax3.imshow(binaryS, cmap='gray')
ax3.set_title('Binary S', fontsize=30)

plt.show()
```

```python
# You can also see that in the H channel, 
# the lane lines appear dark, so we could try a 
# low threshold there and obtain the following result:

thresh = (15, 100)
binaryH = np.zeros_like(H)
binaryH[(H > thresh[0]) & (H <= thresh[1])] = 1
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.imshow(H, cmap='gray')
ax1.set_title('H', fontsize=30)
ax2.imshow(binaryH, cmap='gray')
ax2.set_title('Binary H', fontsize=30)
plt.show()
```

Now you can see that, the S channel is still doing a fairly robust job of picking up the lines under very different color and contrast conditions, while the other selections look messy. You could tweak the thresholds and get closer in the other channels, but the S channel is preferable because it is more robust to changing conditions.

It's worth noting, however, that the R channel still does rather well on the white lines, perhaps even better than the S channel. As with gradients, it's worth considering how you might combine various color thresholds to make the most robust identification of the lines.
